```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

#### Tokenization - Convert paragraph to sentences, focus on words

```python
nltk.download('punkt')
sentences = nltk.sent_tokenize(paragraph)
```

```python
print(sentences)
print(type(sentences))
```

### Cleanup before lemmatize

```python
import re #Regular Expression
corpus = []
for i in range(len(sentences)):
    review= re.sub('[^a-zA-Z]', ' ', sentences[i]) #Use Tab on re.sub()
    review=review.lower() #Lower Case
    #review=review.split() #Split
    corpus.append(review)
    
```

```python
corpus
```

### Stemming

```python
#Print each sentence
for i in corpus:
    print(i)
```

```python
##Print words #Incorrect
for i in corpus:
    for word in i:
        print(word)
```

```python
##Print words
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        print(word)
```
