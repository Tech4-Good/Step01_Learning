```python
import pandas as pd
import folium
```

```python
data_path = "https://data.smartdublin.ie/dataset/33ec9fe2-4957-4e9a-ab55-c5e917c7a9ab/resource/2dec86ed-76ed-47a3-ae28-646db5c5b965/download/dublin.csv"
bike_station_locations=pd.read_csv(data_path)
print(bike_station_locations.shape)
bike_station_locations.head(3)
```

```python
bike_station_locations = bike_station_locations[["Latitude", "Longitude", "Name"]]
print(bike_station_locations.shape)
bike_station_locations.head(3)
```

```python
#pass in the mean of the latitude and longitude coordinates - to centre the map there.
map = folium.Map(location=[bike_station_locations.Latitude.mean(), bike_station_locations.Longitude.mean()], zoom_start=14, control_scale=True)
display(map)
```

```python
greenwich_map =folium.Map(location=[51.477928, -0.001545], zoom_start=16)
display(greenwich_map)  
```

```python
home_map = folium.Map(location=[16.074311, 78.903074], zoom_start=18)
display(home_map)
```

```python
#The iterrows() method generates an iterator object of the DataFrame, allowing us to iterate each row in the DataFrame.
#Each iteration produces an index object and a row object (a Pandas Series object).
#iterrows() returns the index, and the content as a Pandas Series object.

for index, location_info in bike_station_locations.iterrows():
    folium.Marker([location_info["Latitude"], location_info["Longitude"]], popup=location_info["Name"]).add_to(map)
display(map)  
```
